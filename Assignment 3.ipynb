```python
name = input("Enter your name: ")
age = int(input("Enter your age: "))
hobby = input("Enter your hobby: ")

if age <= 12:
    category = "Child"
elif age <= 19:
    category = "Teenager"
elif age <= 59:
    category = "Adult"
else:
    category = "Senior Citizen"

print("\nHello", name + "!")
print("You are classified as:", category)
print("It's great that you enjoy", hobby + "!")
```
