```python
# Simple ML Idea Simulation

# College Prediction Example
attendance = float(input("Enter attendance percentage: "))
internal_marks = float(input("Enter internal marks: "))

if attendance > 75 and internal_marks > 40:
    print("Prediction: Pass")
else:
    print("Prediction: At Risk")

# Healthcare Risk Example
age = int(input("Enter age: "))
bp = int(input("Enter blood pressure: "))

if age > 50 and bp > 140:
    print("Health Risk: High")
else:
    print("Health Risk: Low")

# Shopping Recommendation Example
purchase_count = int(input("Enter number of previous purchases: "))

if purchase_count > 5:
    print("Recommended: Premium Products")
else:
    print("Recommended: Popular Products")
```
