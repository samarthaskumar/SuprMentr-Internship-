```python
import numpy as np

# Given list of temperatures
temp_list = [28, 22, 30, 37, 26, 48]

# Convert list to NumPy array
temp_array = np.array(temp_list)

# Print NumPy array
print("Temperature Array:", temp_array)

# Maximum and Minimum temperature
print("Maximum Temperature:", np.max(temp_array))
print("Minimum Temperature:", np.min(temp_array))

# Average temperature
print("Average Temperature:", np.mean(temp_array))

# Last three days temperature
print("Last Three Days Temperature:", temp_array[-3:])
```
