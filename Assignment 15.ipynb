```python
# Step 1: Import libraries
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# Step 2: Create dataset (emails and labels)
emails = [
    "Win a free phone now",
    "Limited time offer click now",
    "Meeting tomorrow at office",
    "Project discussion today",
    "Congratulations you won prize"
]

labels = ["spam", "spam", "not spam", "not spam", "spam"]

# Step 3: Convert text into numbers
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(emails)

# Step 4: Train the model
model = MultinomialNB()
model.fit(X, labels)

# Step 5: Test with a new email
test_email = ["Free prize waiting for you"]
test_vector = vectorizer.transform(test_email)

prediction = model.predict(test_vector)

print("Prediction:", prediction[0])
```
