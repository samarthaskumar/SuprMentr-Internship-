```python
# Step 1: Import libraries
import numpy as np
from sklearn.linear_model import LinearRegression

# Step 2: Training data (house size vs price)
# size in square feet
size = np.array([500, 700, 900, 1100, 1300]).reshape(-1,1)

# price in thousands
price = np.array([100, 150, 200, 250, 300])

# Step 3: Create model
model = LinearRegression()

# Step 4: Train the model
model.fit(size, price)

# Step 5: Predict price for new house
new_size = np.array([[1000]])
predicted_price = model.predict(new_size)

print("Predicted house price:", predicted_price)
```
