```python
# Step 1: Import required libraries
import numpy as np
import time

# Step 2: Create 1 million numbers using Python list
python_list = list(range(1000000))

# Step 3: Measure time for Python list operation
start = time.time()

new_list = []
for i in python_list:
    new_list.append(i * 2)

end = time.time()
print("Python List Time:", end - start)

# Step 4: Create NumPy array with 1 million numbers
numpy_array = np.arange(1000000)

# Step 5: Measure time for NumPy operation
start = time.time()

new_array = numpy_array * 2

end = time.time()
print("NumPy Array Time:", end - start)
```
