```python
# KNN-based Netflix-like Recommendation (User-Based)

import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Users' movie ratings (rows = users, columns = movies)
# Movies: [Action, Comedy, Drama]
ratings = np.array([
    [5, 3, 0],  # User A
    [4, 2, 0],  # User B
    [1, 5, 4]   # User C
])

# Compute similarity between users
similarity = cosine_similarity(ratings)

print("User Similarity Matrix:\n", similarity)

# Recommend movies for a target user
def recommend_movies(user_index, ratings, similarity, k=1):
    # Get similarity scores for the user
    sim_scores = similarity[user_index]
    
    # Get indices of top-k similar users (excluding itself)
    similar_users = np.argsort(sim_scores)[::-1][1:k+1]
    
    # Weighted sum of ratings from similar users
    weighted_ratings = np.zeros(ratings.shape[1])
    sim_sum = 0
    
    for u in similar_users:
        weighted_ratings += sim_scores[u] * ratings[u]
        sim_sum += sim_scores[u]
    
    # Normalize
    if sim_sum != 0:
        weighted_ratings /= sim_sum
    
    # Recommend movies not yet watched (rating = 0)
    recommendations = []
    for i in range(len(ratings[user_index])):
        if ratings[user_index][i] == 0:
            recommendations.append((i, weighted_ratings[i]))
    
    return sorted(recommendations, key=lambda x: x[1], reverse=True)

# Get recommendations for User A (index 0)
recs = recommend_movies(0, ratings, similarity, k=1)

print("\nRecommended Movies (index, score):")
for movie in recs:
    print(movie)
```
