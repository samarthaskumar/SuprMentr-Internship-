```python
import matplotlib.pyplot as plt

# Data
subjects = ["Python", "Java", "C++", "AI", "ML"]
students = [40, 35, 25, 30, 20]
marks = [60, 70, 75, 80, 85, 90, 95, 70, 65, 88]

# Bar Chart
plt.bar(subjects, students)
plt.title("Students in Each Subject")
plt.show()

# Pie Chart
plt.pie(students, labels=subjects)
plt.title("Subject Distribution")
plt.show()

# Histogram
plt.hist(marks)
plt.title("Marks Distribution")
plt.show()
```
